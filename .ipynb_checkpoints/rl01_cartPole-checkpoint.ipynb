```python
import gym
import matplotlib.pyplot as plt
env = gym.make('CartPole-v1')
num_eps = 1000
steps_total = []
for ep in range(num_eps):
    state = env.reset()
    step = 0
    while True:
        step+=1
#     for step in range(100)

        action = env.action_space.sample()
        new_state, reward, done, info = env.step(action)
#         env.render()
#         print(new_state)
#         print(info)
        if done:
            steps_total.append(step)
            print("Episode finished after steps: %d"%step)
            break;
            
print("Average number of steps: %.2f"%((sum(steps_total))/num_eps))

plt.plot(steps_total)
plt.show()
```
